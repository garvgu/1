### Section 1: Handling Duplicates
##### Task 1: Identify and remove duplicate rows.


```python
# Check number of duplicate rows
duplicate_count = df.duplicated().sum()
print("Number of duplicate rows:", duplicate_count)

# Remove duplicate rows
df = df.drop_duplicates()

# Confirm removal
print("Duplicates after removal:", df.duplicated().sum())
```

### Section 2: Handling Missing Values
##### Task 2: Identify missing values in `CodingActivities`.


```python
# Check missing values
print("Missing before:", df["ConvertedCompYearly"].isna().sum())

# Compute median
median_value = df["ConvertedCompYearly"].median()

# Replace NaN safely
df["ConvertedCompYearly"] = df["ConvertedCompYearly"].fillna(median_value)

print("Missing after:", df["ConvertedCompYearly"].isna().sum())

```

##### Task 3: Impute missing values in CodingActivities with forward-fill.


```python
# Check missing values
print("Missing values before:", df["ConvertedCompYearly"].isna().sum())

# Replace NaN with median
median_value = df["ConvertedCompYearly"].median()
df["ConvertedCompYearly"].fillna(median_value, inplace=True)

print("Missing values after:", df["ConvertedCompYearly"].isna().sum())

```

**Note**:  Before normalizing ConvertedCompYearly, ensure that any missing values (NaN) in this column are handled appropriately. You can choose to either drop the rows containing NaN or replace the missing values with a suitable statistic (e.g., median or mean).
